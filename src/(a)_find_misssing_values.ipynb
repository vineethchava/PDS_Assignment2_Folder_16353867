```python
import pandas as pd
```

```python
data_train_raw = pd.read_csv('/content/train (1).csv')
```

```python
data_train_raw.head()
```

```python
missing_info = data_train_raw.isnull().sum()
print("Missing values in each column:\n", missing_info)
missing_percentage = (data_train_raw.isnull().sum() / len(data_train_raw)) * 100
print("\nPercentage of missing values:\n", missing_percentage)
for column in data_train_raw.columns:
    if data_train_raw[column].isnull().sum() > 0:
        # For numerical columns, using mean or median
        if data_train_raw[column].dtype in ['float64', 'int64']:
            # Checking distribution to decide mean or median
            if data_train_raw[column].skew() < 1:
                data_train_raw[column].fillna(data_train_raw[column].mean(), inplace=True)
                print(f"Filled missing values in '{column}' with mean.")
            else:
                data_train_raw[column].fillna(data_train_raw[column].median(), inplace=True)
                print(f"Filled missing values in '{column}' with median.")
        # For categorical columns, using mode
        else:
            data_train_raw[column].fillna(data_train_raw[column].mode()[0], inplace=True)
            print(f"Filled missing values in '{column}' with mode.")
```
